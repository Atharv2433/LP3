```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("./sales_data_sample.csv", encoding='unicode_escape')
df = data.copy()
```

```python
df.info()
```

```python
# dorpping useless columns
columns = ['ADDRESSLINE1', 'ADDRESSLINE2', 'PHONE', 'STATE', 'POSTALCODE']
df = df.drop(columns, axis=1)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df = df.drop(['TERRITORY'], axis=1)
```

```python
df
```

```python
plt.figure(figsize=(15,5))
sns.boxplot(df)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(df.select_dtypes(['int64', 'float64']))
```

```python
from sklearn.cluster import KMeans
```

```python
inertia = []
k = [i for i in range(1,200)]

for i in k:
    kmeans = KMeans(n_clusters=i, random_state=42, n_init=10)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(15,5))
plt.plot(k, inertia)
plt.grid(True)
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
labels = kmeans.fit_predict(X)

sns.scatterplot(data=df,x='SALES',y='MSRP',hue=labels)
```
